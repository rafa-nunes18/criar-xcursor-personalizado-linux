#### Como criar o cursor de mouse no linux mint 22

### Passo 01 - Criando o arquivo de imagem

A primeira coisa que você precisará é criar seu próprio arquivo de imagem PNG.

Você deve usar imagens quadradas com uma das seguintes resoluções: 16x16, 22x22, 32x32, 48x48, 64x64, 128x128 256x256.

Certifique-se de manter a camada de transparência (informação alfa) no seu arquivo ao salvar.

Recomenda-se salvar o arquivo com um nome que tornará mais fácil para você inferir seu uso. A estrutura deve seguir esse padrão: filename.png.
Exemplo : 

    default.png.

### Passo 02 - Determinando o ponto de acesso ao Cursor (hotspot).

O ponto de acesso de um cursor é o ponto em que o clique ocorre. É a posição na sua imagem que o sistema irá interpretar e entender como a posição do cursor.

Para determinar o pixel onde o ponto de acesso deve ficar, é necessário usar edição gráfica software capaz de determinar o pixel exato para o hotspot.

O ponto de origem (0,0) está no canto superior esquerdo da imagem. O eixo X vai da esquerda para a direita. O eixo Y vai de cima para baixo.

Com isso em mente, quando você adiciona um deslocamento de 10 pixels no eixo X, ele mudará o ponto de acesso 10 pixels para a direita a partir do canto superior esquerdo. Da mesma forma, um deslocamento de 5 pixels no eixo Y Ele vai mudar o hotspot de 5 pixels para baixo do canto superior esquerdo.
Importante!

### Passo 03 - Criando o arquivo .cursor

Agora você sabe o htspot e tem a imagem, hora de criar o seu .cursor do ficheiro. que é um arquivo de configuração que irá dizer ao xcursorgen como gerar o cursor.

É estruturado da seguinte forma: (resolution) (hotspot-x) (hotspot-y) (filename) 

Crie com qualquer programa que lhe permita salvar o arquivo com uma extensão personalizada.
Neste exemplo, usar o echo comando em um terminal.

    echo "32 10 5 default.png" > default.cursor

32 é para a resolução, então isso representa um arquivo de 32x32 pixels.
10 é para a coordenada hotspot-x, então isso representa um deslocamento de 10 pixels no eixo X.
5 é para a coordenada hotspot-y, então isso representa um deslocamento de 5 pixels no eixo Y.
O nome do arquivo de imagem correspondente, incluindo sua extensão.
">" é para o comando eco despejar o resultado no seguinte destino.
default.cursor é o arquivo de destino que receberá a saída do comando eco.

### Passo 04 - Usando o xcursorgen

Usar o xcursorgen comando para gerar o cursor, abra um terminal e navegue até o diretório onde os arquivos de imagem do cursor estão armazenados.

O xcursorgen trabalha na seguinte estrutura: (xcursorgen) (config-file) (output-file).
exemplo:

    xcursorgen default.cursor default

Ele irá ler a configuração a partir do default.cursor arquivo e gerar Um novo cursor chamado default no seu diretório de trabalho.
